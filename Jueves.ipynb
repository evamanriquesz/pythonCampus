# Dataframes 

https://pandas.pydata.org/docs/user_guide/index.html

```python
import pandas as pd

# Crear un diccionario con los datos
data = {
    'Nombre': ['Pelota', 'Mu√±eca', 'Cami√≥n', 'Rompecabezas', 'Lego'],
    'Tipo': ['Deporte', 'Figura', 'Veh√≠culo', 'Juego de mesa', 'Construcci√≥n'],
    'Precio': [5, 10, 7, 12, 15],
    'Stock': [50, 30, 20, 10, 15]
}

# Crear el dataframe
df = pd.DataFrame(data)

# Mostrar el dataframe
print(df)
```

```python
df
```

```python
# En un df largo, para ver las primeras filas: 
df.head()
```

```python
#Seleccionar una columna 
print(df['Nombre'])
```

```python
#Filtrar por una condicion
productos_caros = df[df['Precio'] > 10]
print(productos_caros)
```

```python
# A√±adir una nueva columna 
df['Precio con IVA'] = df['Precio'] * 1.21
print(df)
```

```python
# Ordenar por una columna
df_sorted = df.sort_values(by='Precio')
print(df_sorted)
```

```python
# Agrupar por 'Tipo' y calcular la media del 'Precio'
grouped = df.groupby('Tipo')['Precio'].mean()
print(grouped)
```
