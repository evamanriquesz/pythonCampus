```python
import matplotlib.pyplot as plt
```

```python
# Grafico de lineas
```

```python
x = [2, 3, 5, 6, 8]
y = [1, 5, 6, 8, 10]
```

```python
plt.plot(x, y)
plt.grid()
```

```python
plt.plot(x, y, color ="green")
```

```python
plt.plot(x, y, color ="green", linestyle="--")
```

```python
plt.plot(x, y, color ="green", linestyle="--", linewidth =5)
```

```python
x = [2, 3, 5, 6, 8]
y = [1, 5, 6, 8, 10]

x2 = [1,4, 5, 7,8]
y2 = [7, 12, 15, 17, 20]

plt.plot(x, y, color ="green", linestyle="--", linewidth =5)
plt.plot(x2, y2, color ="red", linestyle="--", linewidth =2)
```

```python
plt.plot(x, y, color ="green", linestyle="--", linewidth =5, label="Clase1")
plt.plot(x2, y2, color ="red", linestyle="--", linewidth =2, label="Clase2")
plt.legend(loc = "upper left")
plt.title("Grafico de lineas de prueba")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
```
